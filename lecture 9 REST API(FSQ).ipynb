```python
%%writefile place_search.py

import requests

def place_search(querry,lat,lon,sort='RELEVANCE',limit=10,field="POI - Core Data"):
    '''
    querry: Required argument
    ll: Required argument
    sort:optional argument
         by default : RELEVANCE
         option: RATING,DISTANCE,POPULARITY
    limit:
        by default 10
        max_limit 50
    filed=POI - Core Data,POI - Rich Data,Address Data

    '''
    api_key='fsq31a/KD/geG2lMsLoa9J5SytpoYUMalZfqLxSVlkn0fao='

    url = "https://api.foursquare.com/v3/places/search"
    ll=f"{lat},{lon}"
    params = {
        "query": querry,
        "ll": ll,
        "open_now": "true",
        "sort":sort,
        "filed":field
    }

    headers = {
        "Accept": "application/json",
        "Authorization": f"{api_key}"
    }

    response = requests.get( url, params=params, headers=headers)

    if response.status_code==200:
        data=response.json()
        for i in data['results']:
            name=i['name']
            fsq_id=i['fsq_id']
            print(fsq_id)
            print(name)
            print('-'*100)
            print('-'*100)
            print()
            
    else:
        print(f'error occured status code:{response.status_code} reason:{response.reason}')
```
